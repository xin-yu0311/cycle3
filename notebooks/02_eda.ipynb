```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from pathlib import Path

ROOT = Path('..')
DATA_RAW = ROOT / 'data' / 'raw' / 'YRBS_2007.csv'

recoded = pd.read_csv(ROOT / 'data' / 'processed' / 'yrbs_recoded.csv')
recoded.head()
```

```python
summary = recoded.groupby('Sex')['SadOrHopeless_binary'].agg(sample_size='count', yes_count='sum').reset_index()
summary['no_count'] = summary['sample_size'] - summary['yes_count']
summary['sample_proportion'] = summary['yes_count'] / summary['sample_size']
summary['percent'] = summary['sample_proportion'] * 100
summary['Sex'] = pd.Categorical(summary['Sex'], ['Female','Male'], ordered=True)
summary = summary.sort_values('Sex').astype({'Sex':'str'}).reset_index(drop=True)
summary.to_csv(ROOT / 'outputs' / 'tables' / 'eda_summary_table.csv', index=False)
summary
```

```python
cross = pd.crosstab(recoded['Sex'], recoded['SadOrHopeless_binary'], normalize='index').rename(columns={0:'No',1:'Yes'}).reset_index()
cross.to_csv(ROOT / 'outputs' / 'tables' / 'response_proportion_by_group.csv', index=False)
cross
```

```python
plt.figure(figsize=(7,4.5))
plt.bar(summary['Sex'], summary['percent'])
plt.ylabel('Percent reporting sad or hopeless feelings (%)')
plt.xlabel('Sex')
plt.title('Sad or Hopeless Feeling by Sex')
for i, row in summary.iterrows():
    plt.text(i, row['percent']+0.8, f"{row['percent']:.2f}%", ha='center')
plt.ylim(0, max(summary['percent'])+10)
plt.tight_layout()
plt.savefig(ROOT / 'outputs' / 'figures' / 'sad_hopeless_by_sex.png', dpi=200)
plt.show()
```
