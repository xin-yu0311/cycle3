# 01 Data Check

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from pathlib import Path

ROOT = Path('..')
DATA_RAW = ROOT / 'data' / 'raw' / 'YRBS_2007.csv'

df = pd.read_csv(DATA_RAW)
df.head()
```

```python
df[['WhatIsYourSex', 'SadOrHopeless']].describe()
```

```python
selected = df[['WhatIsYourSex', 'SadOrHopeless']].copy()
for col in selected.columns:
    selected[col] = pd.to_numeric(selected[col], errors='coerce')
cleaned = selected[selected['WhatIsYourSex'].isin([1,2]) & selected['SadOrHopeless'].isin([1,2])].copy()
print('Raw rows:', len(df))
print('Cleaned valid rows:', len(cleaned))
cleaned.to_csv(ROOT / 'data' / 'processed' / 'yrbs_cleaned.csv', index=False)
cleaned.head()
```

```python
recoded = cleaned.copy()
recoded['Sex'] = recoded['WhatIsYourSex'].map({1:'Female', 2:'Male'})
recoded['SadOrHopeless_binary'] = recoded['SadOrHopeless'].map({1:1, 2:0})
recoded = recoded[['WhatIsYourSex','Sex','SadOrHopeless','SadOrHopeless_binary']]
recoded.to_csv(ROOT / 'data' / 'processed' / 'yrbs_recoded.csv', index=False)
recoded.head()
```
